```python
def custom_map(x):
    # 定义映射参数，可根据实际情况调整以改变映射效果
    max_mapped_value = 3000  # 映射后的最大值
    min_mapped_value = 0     # 映射后的最小值
    transition_point = 20    # 前后过渡点
    linear_slope = (max_mapped_value - min_mapped_value) / transition_point
    exponential_decay_rate = 1.5  # 控制指数衰减速度

    if x <= transition_point:
        # 线性增长部分
        mapped_value = min_mapped_value + linear_slope * x
    else:
        # 指数递减部分
        mapped_value = max_mapped_value - (max_mapped_value - min_mapped_value) * math.exp(-exponential_decay_rate * (x - transition_point))

    return max(min_mapped_value, min(mapped_value, max_mapped_value))

# 对自然数范围进行映射
for i in range(30):
    mapped_i = custom_map(i)
    plt.scatter(i,mapped_i)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def exponential_func(x, A=90000, B=0.1485, C=np.log(3000)):
    y = A * np.exp(B * x - C)
    return y

# 设置定义域
x_values = np.linspace(0, 30, 1000)

# 计算对应的函数值
y_values = exponential_func(x_values)

# 绘制函数图像
plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values)
# plt.title('y=145000*exp(0.04-log3000)')
plt.title('y=90000*exp(0.1485-log3000)')
plt.xlabel('x (Domain: 0 to 30)')
plt.ylabel('y (Range: approximately 0 to 3000)')
plt.grid(True)
plt.xlim(0, 30)
plt.ylim(0, 3000)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def exponential_func(x, A=90000, B=0.1485, C=np.log(3000)):
    y = A * np.exp(B * x - C)
    return y
def linear_func(x, k=100):
    y = k*x
    return y

# 设置定义域
x_values = np.linspace(0, 30, 1000)

# 计算对应的函数值
y_values = linear_func(x_values)

# 绘制函数图像
plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values)
# plt.title('y=145000*exp(0.04-log3000)')
plt.title('y=100x')
plt.xlabel('x (Domain: 0 to 30)')
plt.ylabel('y (Range: approximately 0 to 3000)')
plt.grid(True)
plt.xlim(0, 30)
plt.ylim(0, 3000)
plt.show()
```
